```python
sp_500_index = '^GSPC'
tickers_merged = russell_1000_tickers + [sp_500_index]
```

```python
data = yf.download(tickers_merged, start = '2021-01-01', end = '2025-05-30', interval = '1d', auto_adjust= 'False')
```

```python
import holidays
us_holidays = holidays.UnitedStates(years=[2025,2024,2023,2022, 2021])
date_range = pd.date_range(start='2021-01-01', end='2025-05-30', freq='B')
holiday_dates = pd.to_datetime([date for date in us_holidays])
trading_days = date_range[~date_range.isin(holiday_dates)]
filtered_data = data.loc[data.index.isin(trading_days)]
```

```python
df_nefiltrirani = data.stack(level=1)
df_filtrirani = filtered_data.stack(level = 1)
```

```python
filtered_ticker = df_filtrirani.loc[df_filtrirani['Ticker'] == 'AAPL']
filtered_ticker
#vraćam se na ovaj dio jer se ne može napraviti shift za open value bez da se to uradi za isti ticker, na to treba obratiti pažnju!
#uzet je kao primjer Apple, treba uraditi shift koji je planiran za kasnije labeliranje na nacin da se dodijeli naredna vrijednost za opening value ali za ISTI ticker
```

```python
df_filtrirani["Open tomorrow"] = df_filtrirani.groupby(["Ticker"])["Open"].shift(-1)
```

Šta je do sada urađeno : skinuti su tickeri sa Wikipedija stranica za Russell 1000 univerzum, dodan ticker za S&P 500. Prebačeno je u csv i potom u Pandas dataframe. Pri skidanju iskorišten je auto_adjust = false, pa je dodana kolona Adj close u dataframe. Ta kolona neće biti korištena, pa je uklonjena u narednim koracima sa drop funkcijom u Pandasu. Namještene su kolone da su sve u istom nivou pa idući korak koji slijedi jeste čišćenje dataseta od podataka koji nisu relevantni (informacije za non-working days). Uzet će se aproksimacija podataka tako da se uklone vikendi i praznici berze u Americi. NOTE : Ipak je urađeno čišćenje podataka prije prebacivanjau csv i to putem biblioteke holidays

DODATNO: Dodana je kolona Open tomorrow koja označava opening price za sutrašnji dan. Ta kolona je morala biti napravljena opreznim shiftom, dakle, moralo je biti urađeno grupisanje prema tickeru, te potom shift za -1. U nastavku se filtrira opet ticker AAPL kako bi se potvrdilo da je zaista urađen shift prema onome što je navedeno

```python
# sad se moze napraviti i target na osnovu prethodno dodane kolone
#napravit cu dvije dodatne kolone, jedna da bude Return value, druga da bude Return target
df_filtrirani["Return value"] = ((df_filtrirani["Open tomorrow"] - df_filtrirani["Open"])/ df_filtrirani["Open"])
df_filtrirani["Return target"] = (abs(df_filtrirani["Return value"]) == df_filtrirani["Return value"])
```

```python
df_filtrirani["Return target"]= df_filtrirani["Return target"].astype(int)
```

```python
df_filtrirani
#trenutno je jedini problem sto je prikazana 0 za value koji ne postoji uopće, a to je Return target za 29.05
```

```python
filtered_date = df_filtrirani.loc[df_filtrirani['Date'] == '2025-05-29']
filtered_date
```

U ovom dijelu vidi se da je problematično to što su za Return target dodijeljene vrijednosti 0, bez obzira na to što ta vrijednost nije poznata (vidimo iz toga što je return value NaN, kao i Open tomorrow). Jasno je da i Return target treba biti NaN, pa će to biti i urađeno u narednom koraku

```python
import numpy as np
mask = (df_filtrirani['Date'] == '2025-05-29')
df_filtrirani.loc[mask, 'Return target'] = np.nan
```

```python
filtered_date = df_filtrirani.loc[df_filtrirani['Date'] == '2025-05-29']
filtered_date
```

ML dio - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100, min_samples_split= 50, random_state = 1)
train_data = df_filtrirani.iloc[:-200000]
test_data = df_filtrirani.iloc[-200000:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model_rf.fit(train_data[predictors], train_data["Return target"])
```

```python
#trebat ce se ponovo definisati drugacije test data, ali da ne bih trenirala nanovo model sad, ovdje cu ga ponovo redefinirati
#stvar je u tome da se trebaju ukloniti sve NaN vrijednosti iz testnog dataseta jer nema smisla da budu tu, pravi se error kod metrika
#uklanjam sve redove koji sadrze bilo kakvu NaN vrijednost
test_data = test_data.dropna()
test_data
```

```python
from sklearn.metrics import precision_score
predictions = model_rf.predict(test_data[predictors])
```

```python
predictions= pd.Series(predictions, index = test_data.index)
```

```python
predictions
```

U suštini ovo govori da li stvarno cijena dionica raste u trenutku kad je predikcija rekla da će rasti, kao i obrnuto

```python
precision_modela = precision_score(test_data["Return target"], predictions)
precision_modela
```

```python
combined_test_and_pred = pd.concat([test_data["Date"], test_data["Return target"], predictions], axis = 1 )
```

```python
combined_test_and_pred.rename(columns={ 0 : 'Prediction return'}, inplace = True)
combined_test_and_pred
#Poslije cu se vratiti na to da plottam jednostavno ovo, sad je bolje uraditi backtesting
```

```python
combined_test_and_pred.to_csv('Combined_test_return_and_predictions_return.csv')
```

Sad bi trebalo uraditi backtestiranje

```python
#Recall - number of true positives in the model; considered good for 0.75 or above
from sklearn.metrics import recall_score
recall_modela = recall_score(combined_test_and_pred["Return target"], combined_test_and_pred["Prediction return"])
```

```python
recall_modela
```

```python
F1 = 2* (precision_modela * recall_modela) / (precision_modela + recall_modela)
F1
```
